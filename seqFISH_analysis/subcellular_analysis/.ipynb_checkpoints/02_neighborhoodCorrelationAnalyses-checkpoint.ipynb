# Explode paired correlation

```python
def row_col_value(df, pairs):
    corr = []
    for r_c in pairs:
        r,c = r_c.split('_')
        corr.append(df.loc[r, c])
        if not(df.loc[r,c] == df.loc[r,c]):
            print(df)
    return np.array(corr)
```

```python
# Find marker pairs
markers = bm_exp_df.columns.tolist()
pairs = []
for i in range(len(markers)):
    for j in range(len(markers)):
        if not ((markers[i]+'_'+markers[j] in pairs) or (markers[j]+'_'+markers[i] in pairs)):
            pairs.append(markers[i]+'_'+markers[j])
```

```python
bm_corr_exploded_l = []
uc_corr_exploded_l = []
hch_corr_exploded_l = []
for df in bm_corr_l:
    bm_corr_exploded_l.append(row_col_value(df, pairs))
for df in uc_corr_l:
    uc_corr_exploded_l.append(row_col_value(df, pairs))
for df in hch_corr_l:
    hch_corr_exploded_l.append(row_col_value(df, pairs))
bm_corrs = np.vstack(bm_corr_exploded_l)
uc_corrs = np.vstack(uc_corr_exploded_l)
hch_corrs = np.vstack(hch_corr_exploded_l)
```

```python
np.concatenate((bm_corrs, uc_corrs, hch_corrs)).shape
```

```python
z_data = np.concatenate((bm_corrs, uc_corrs, hch_corrs))
z_data = (z_data - np.mean(z_data, axis=0))/np.std(z_data,axis=0)
z_data = np.nan_to_num(z_data)
```
