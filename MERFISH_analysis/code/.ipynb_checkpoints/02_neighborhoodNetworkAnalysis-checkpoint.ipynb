## Batch analysis

```python
adata_dir = r'..\merfish_data\bento_u2os\patch_adata'
out_dir = r'..\merfish_data\140gene_analysis\networks\correlation'
perm_dir = r'..\merfish_data\bento_u2os\networks\perm_interaction'
fn_l = os.listdir(adata_dir)
fn_l.sort()
```

### Permutation

```python
by_cell_groups = data.groupby('cell')
```

```python
for fn in tqdm(fn_l):
    
    adata = anndata.read_h5ad(filename=os.path.join(adata_dir,fn))
    name = fn.split('_')[2]
    
    
    # cellID = name.lstrip('0')
    # if len(cellID)==0:
    #     cellID = '0'
    cellID = name
    
    patch = adata.obs['leiden'].tolist()
    patch = [int(item) for item in patch]
    
    cell = by_cell_groups.get_group(cellID)
    cell.insert(cell.shape[1], 'patch', patch)
    
    patch_l = np.unique(cell['patch'].tolist())
    patch_l.sort()
    interaction_l = Parallel(n_jobs=8)(delayed(nha.permutation)(cell.loc[cell['patch']==p], n_neighbors=10,
                                     row_name='y',col_name='x',
                                      marker_names='gene',n_permutation=1000) for p in patch_l)
    
    name = 'u2os_'+name
    with open(os.path.join(perm_dir, name+'_interactions_10_neighbors.pkl'), 'wb') as f:
        pickle.dump(interaction_l, f)
```

### Correlation

```python
for fn in tqdm(fn_l):
    
    adata = anndata.read_h5ad(filename=os.path.join(adata_dir, fn))
    name = fn.split('_')[2]
    
    cellID = name.lstrip('0')
    if len(cellID) == 0:
        cellID = '0'
    cellID = int(cellID)
    
    patch = adata.obs['leiden'].tolist()
    patch = [int(item) for item in patch]
    
    cell = bycell_groups.get_group(cellID)
    cell.insert(8, 'patch', patch)
    
    corr_l = []
    for i in range(len(np.unique(cell['patch'].tolist()))):
        corr_l.append((proximity_network(cell.loc[cell['patch']==i],n_neighbors=10,
                                        row_name='wordCentroidY',col_name='wordCentroidX',
                                        marker_names='geneName'
                                       )))
    
    name = 'merfish_cell_'+name
    with open(os.path.join(out_dir, name+'_neighborhoods_10_neighbors.pkl'), 'wb') as f:
        pickle.dump(corr_l, f)
```

```python
# Compute z-scores of each patch
cluster_dir = r'..\merfish_data\140gene_analysis\patch_corr'
out_dir = r'..\merfish_data\140gene_analysis\networks\z_score'
fn_l = os.listdir(cluster_dir)
fn_l.sort()

for fn in fn_l:
    if fn.endswith('.pkl'):
        cell_id = fn.split('_')[2]
        patch_expression = pd.read_pickle(os.path.join(cluster_dir, fn))['expression']
        z_score = (patch_expression-patch_expression.mean())/patch_expression.std()
        with open(os.path.join(out_dir, 'merfish_cell_'+cell_id+'_patch_z-score.pkl'), 'wb') as f:
            pickle.dump(z_score, f)
```

## Network variability

```python
data = pd.read_csv(r'..\merfish_data\MERFISH_Examples\MERFISH_Examples\MERFISH_Examples\analyzed_data\cleaned_data.csv')
```

```python
markers = np.unique(data['geneName'])
pairs = []
for i in range(len(markers)):
    for j in range(len(markers)):
        if not ((markers[i]+'_'+markers[j] in pairs) or (markers[j]+'_'+markers[i] in pairs)):
            pairs.append(markers[i]+'_'+markers[j])
```

```python
cell_ids = data['cell'].tolist()
```

```python
in_dir = r'..\merfish_data\bento_u2os\networks\perm_interaction'
fn_l = os.listdir(in_dir)
var_names = []
arr = []
for fn in tqdm(fn_l):
    l = pd.read_pickle(os.path.join(in_dir, fn))
    temp = []
    m = np.zeros(len(pairs))
    s = np.zeros(len(pairs))
    for j in range(len(pairs)):
        gene1,gene2 = pairs[j].split('_')
        temp.append([])
        for i in range(len(l)):
            if l[i] is None:
                continue
            if (gene1 in l[i][0].columns.tolist()) and (gene2 in l[i][0].columns.tolist()):
                temp[j].append(l[i][0].loc[gene1,gene2])
        if len(temp[j])!=0:
            m[j] = np.mean(np.array(temp[j]))
            s[j] = np.std(np.array(temp[j]))
    arr.append(np.concatenate((m,s)))
```

```python
arr = np.array(arr)
var_names = []
for p in pairs:
    var_names.append(p+'_mean')
    var_names.append(p+'_std')
df = pd.DataFrame(arr, columns=var_names)
```

```python
df
```

```python
out_dir = r'..\merfish_data\bento_u2os\networks'
with open(os.path.join(out_dir, 'merfish_u2os_interaction_network_variability.pkl'), 'wb') as f:
            pickle.dump(df, f)
```
