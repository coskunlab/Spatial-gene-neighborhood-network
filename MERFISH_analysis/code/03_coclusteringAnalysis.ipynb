### 0-padding

```python
def row_col_value(df, pairs):
    corr = []
    for r_c in pairs:
        r,c = r_c.split('_')
        if (not r in df.columns.tolist()) or (not c in df.columns.tolist()):
            corr.append(0)
        else:
            corr.append(df.loc[r, c])
    return np.array(corr)
```

### Load fibroblast correlation

```python
in_dir = r'..\merfish_data\140gene_example_analysis\patch_corr'
fn_l = os.listdir(in_dir)
fn_l.sort()
```

```python
markers = exp.columns.tolist()
pairs = []
for i in range(len(markers)):
    for j in range(len(markers)):
        if not ((markers[i]+'_'+markers[j] in pairs) or (markers[j]+'_'+markers[i] in pairs)):
            pairs.append(markers[i]+'_'+markers[j])
```

```python
corr_l = []
for fn in fn_l:
    corr_df = pd.read_pickle(os.path.join(in_dir, fn))['correlation']
    corr_l.append(row_col_value(corr_df, pairs))
```

```python
corrs = np.vstack(corr_l)
```

### Load u2os cells

```python
in_dir = r'..\merfish_data\bento_u2os\patch_corr'
fn_l = os.listdir(in_dir)
fn_l.sort()
```

```python
markers = u2os_exp.columns.tolist()
pairs = []
for i in range(len(markers)):
    for j in range(len(markers)):
        if not ((markers[i]+'_'+markers[j] in pairs) or (markers[j]+'_'+markers[i] in pairs)):
            pairs.append(markers[i]+'_'+markers[j])
```

```python
u2os_corr_l = []
for fn in fn_l:
    corr_df = pd.read_pickle(os.path.join(in_dir, fn))['correlation']
    u2os_corr_l.append(row_col_value(corr_df, pairs))
```
